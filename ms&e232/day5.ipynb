#### Bertrand Duopoly

- We assume two firms (N=2)
- Each firm chooses a price p<sub>i</sub> greater than 0
- Consumers buy only from firm with lowest price
- Profit of each firm is : 
    u<sub>i</sub>(p<sub>1</sub>,p<sub>2</sub>) = (p<sub>i</sub>-c<sub>i</sub>)*D<sub>i</sub>*(p<sub>1</sub>,p<sub>2</sub>)

If we take the game in which gas stations compete to lower prices and attract demand, marginal cost c = 3. Both gas stations <b>continously lower prices</b> until they are just above 3 (assuming prices are all real values).   

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(range(10))
y = 3*x - 4

plt.plot(x, y)

plt.show()
```

### Brauess's Paradox

We considered the road network and Brauss's paradox, trying to figure out what the best responses are. Similar to Prof Ugander's class, we determined that if our payoff had a particular equation that becomes more expensive.

This is a form of tragedy of the commons. Any shared resource 

```python
import networkx as nx

na = 1.5
nb = 2.5

G = nx.Graph()
G.add_edge("initial", "B", weight=50)
G.add_edge("initial", "A", weight=10*na)
G.add_edge("A", "final", weight=50)
G.add_edge("B", "final", weight=10*nb)
nx.draw(G, with_labels=True)
```
