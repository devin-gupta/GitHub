### Static Games with Complete Information

##### Lecture 1
<i> Static Game </i>: every player simaltenously chooses a strategy. 
<i> Normal form </i>: describe payoffs per player based on actions (distinct from strategy profile). 
<i> Utility function </i>: inputs strategy profile (vector) and outputs scalar quantity of utility.

<b>Coordination Game</b>: every joins an online gaming platform and utlity is based on the number of other players choosing that platform. 
<b>Matching Pennies</b>: both players choose heads or tails. <i>zero-sum game</i> is when all utility adds up to zero.

<i> Weakly dominant </i>: no matter what everyone else plays, this strategy is <u>equal or better</u> than another strategy. 
<i> Strictly dominant </i>: no matter what everyone else plays, this strategy is <u>better</u> than another strategy. 

Dominant strategy equilibrium can be found when a pair of dominant strategies are found and can be mapped to each other. Ex. In prisoner's dilemna, defect-defect is a dominant strategy equilibrium.

Important: bidding your value in a second-price auction is a weakly dominant strategy.


##### Lecture 2
If there isn't a dominant strategy equilibrium, we can use Iterated Elimination of Strictly Dominated Strategies (IESDS). 
where we remove all strictly dominated strategies. We call the game <i> dominance solvable</i> if IESDS returns only 1 strategy profile.

<b>Beauty Contest</b>: each player chooses a number and players want to guess 2/3 of average of other player's guesses. 

If we consider Iterated Elimination of Weakly Dominated Strategies, order matters hence we might end up in different solutions based on the order of elimination. 

<b> Cournot Duopoly </b>: each firm chooses a quantity, based on a marginal cost and demand curve that decreases as production increases. 

The Cournot Duopoly is dominance solvable for 2 players.

<i> Best Response </i>: returns the set of optimal strategies for a player, for every action of every other player. 

##### Lecture 3
<i> Pure Nash Equilibrium </i>: when our strategy is a best response to every other player's strategy. Is robust to unilateral deviations. 

If a game is dominance solvable, the solution is a unqiue pure NE.

There can be no, one, or multiple pure NE. 

<b> Bertrand Duopoly </b>: each firm chooses a price, based on a marginal cost and demand where everyone buys from lowest price.
Bertrand duopoly only has one pure NE, where prices = marginal cost if both marginal costs are same. If different, no pure NE

Network routing games... pareto efficiency is when no player can become better without making other players worse off.

We can define a 'potential function' in network routing game, and when minimized, the potential function is pure NE. 

<i> Best Response Dynamics </i>: players seuqentially choose best response to current strategy of others. for most games they don't converge, but when they do it's a pure NE. 

##### Lecture 4
<i> Mixed Strategies </i>: when players randomize over pure strategies
for a mixed strategy, the new strategy is the probability distribution over the strategy set. 
To find solution, calculate frequency of either strategy p & q such that the expected payoff for opposing player is even no matter what strategy they chose. 

John Nash showed that every game with finite strategies always has mixed or pure nash. (Pure strategies are mixed strategies hence PNE is MNE)

### Dynamic Games with Complete Information
##### Lecture 5

In dynamic games, the rules dictate, when players play and what they know when they play. 

Hence we can draw games in extensive form (game tree) where the nodes indicate the player and the edges indicate the action. Each end node/leaf has the payoffs. And we represent "simaltaneous moves" with a dashed line.

<i> Perfect Information </i>: if all players can observe the entire history of the game. (no dashed lines)
Hence strategies in dynamic games are "complete contingent plans". 

<i> Subgame </i>: subgames are the game played restricted to some parent node. 
<i> Subgame NE </i>: the NE of the subgame based on some strategy profile.

You can have NE that are not SPNE. 

<i> backward induction </i>: starting at the last decision, work the decision tree back upwards to find the equilibrium path of play.

<b> centipede game </b>: players can continue or drop, will make more and more money if they drop on their turn but doubling incentive for staying. 

<i> Zermelo's Thereom </i>: for any two player game with alternating turns, can guarantee that someone wins or draw - there is an optimal strategy. 

##### Lecture 6

<i> Stackelberg Competition </i>: one firm moves firm, all others follow. everyone chooses quantity with marginal cost and inverse demand curve. 

Sequential bargaining versus stackelberg competition. the first confers last mover advantage while the first confers first-mover advantage.

<i> rubenstein bargaining </i>: sequential bargaining with discount factors in infinite time horizon.
